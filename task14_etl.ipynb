```python
import os

os.makedirs("data/raw", exist_ok=True)
os.makedirs("data/processed", exist_ok=True)
os.makedirs("data/output", exist_ok=True)

print("Folders created successfully!")
```

```python
import pandas as pd

df = pd.read_csv("customer_churn.csv")   # change file name
df.head()
```

```python
df.info()
df.shape
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
df.dtypes
```

```python
df['high_value_customer'] = df['total_spend'].apply(lambda x: 1 if x > 500 else 0)
```

```python
df['long_tenure'] = df['tenure'].apply(lambda x: 1 if x > 12 else 0)
```

```python
customers = df[['customer_id','age','gender','subscription_type','contract_length']]
```

```python
df.columns = df.columns.str.replace("__", "_")
df.columns
```

```python
usage = df[['customer_id','tenure','usage_frequency','support_calls','payment_delay','last_interaction']]
```

```python
churn = df[['customer_id','total_spend','churn','high_value_customer','long_tenure']]
```

```python
import sqlite3

conn = sqlite3.connect("database.sqlite")

customers.to_sql("customers", conn, if_exists="replace", index=False)
usage.to_sql("usage", conn, if_exists="replace", index=False)
churn.to_sql("churn", conn, if_exists="replace", index=False)

conn.close()

print("Data loaded successfully!")
```

```python
df.to_csv("processed_data.csv", index=False)
```
